```python
from lxml import etree
from collections import defaultdict
import urllib.parse

# nltk.download('punkt')
# nltk.download('stopwords')

STOP_WORDS = set(stopwords.words("english"))
set.update({"program", "undergraduate", "students","the"})

url_lst = []

home_url = "https://catalogue.uci.edu/undergraduatedegrees/#:~:text=MINORS%20(organized%20alphabetically)-,UNDERGRADUATE,-MAJORS%20AND%20MINORS"

soup = BeautifulSoup(requests.get(home_url).content, 'lxml')

for link in soup.find_all('a'):
    url_lst.append(link.get('href'))

# print(url_lst.index('#'))
new_url_lst = url_lst[25:]


new_recent_url_lst = new_url_lst[39:]

url_lst = new_recent_url_lst[:424]

new_combined_urls = [urllib.parse.urljoin(home_url, url) for url in url_lst if 'minor' not in url]

# counter = 1
major_tokens_map = defaultdict(list)

for url in new_combined_urls:
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'lxml')
    print('processing html')
    print(url.split('/')[-2])
    

    try:
        for content in soup.select("[id^='textcontainer'] p"):
            tokens = nltk.word_tokenize(content.text)
            tokens_filtered = [word.lower() for word in tokens if word.isalpha() and word.isascii() and word
                             not in STOP_WORDS]
            major_tokens_map[url.split('/')[-2]] = tokens_filtered   
    except AttributeError:
          print("No text")
print(major_tokens_map)

```
